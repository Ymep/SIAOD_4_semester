Задание

Реализовать заданный метод сортировки строк числовой матрицы в соответствии с индивидуальным заданием. Оценить время работы каждого алгоритма сортировки и сравнить его со временем стандартной функции сортировки, используемой в выбранном языке программирования.

Метод:
- Вставкой

```
import random
import timeit
def random_arr(length=100):  # генерирует рандомный массив массив
    arr = []
    while len(arr) < length:
        arr.append(random.randint(-1000, 1000))
    return arr
init_arr = random_arr(100)  # Генерация массива рандомного массива из 1000 элементов
print(init_arr) # выводим для наглядности
def default_sort(arr):  # дефолтная сортировка через sort
    arr.sort()
    return arr

def vs_sort(arr):  # Сортировка вставками
    for index in range(1, len(arr)):
        x = arr[index]
        i = index -1
        while i>=0:
            if x < arr[i]:
                arr[i+1] = arr[i]
                arr[i] = x
                i = i - 1
            else:
                break
    return arr

def qsort(array): # быстрая сортировка
    if not array:  # Если пустой массив
        return array
    pivot = array[random.choice(range(0, len(array)))]

    head = qsort([x for x in array if x < pivot])
    tail = qsort([x for x in array if x > pivot])
    return head + [x for x in array if x == pivot] + tail
# Сортировка массива и вывод времени
print(vs_sort(init_arr))
print("Сортировка по умолчанию   " + str(timeit.timeit("default_sort(init_arr)", number=1, globals=globals())))
print("Быстрая сортировка        " + str(timeit.timeit("qsort(init_arr)", number=1, globals=globals())))
print("Сортировка вставками      " + str(timeit.timeit("vs_sort(init_arr)", number=1, globals=globals())))
```
